```python
import requests
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import CountVectorizer
import pickle
import re
import numpy as np
```

```python
url=input("Enter url")
page=requests.get(url)
```

## 

```python
def cleannn(text):
    cleaned = re.compile(r'<.*?>')
    return re.sub(cleaned,'',text)

def is_special(text):
    rem = ''
    for i in text:
        if i.isalnum():
            rem = rem + i
        else:
            rem = rem + ' '
    return rem

def to_lower(text):
    return text.lower()

def rem_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    return [w for w in words if w not in stop_words]

def stem_txt(text):
    ss = SnowballStemmer('english')
    return " ".join([ss.stem(w) for w in text])
```

```python
txt = cleantext.split(".,")
txt[0]
```

```python
cv = CountVectorizer(max_features = 1000)
```

```python
finaltxt=[]
for review in txt:
    f1=cleannn(review)
    f2=is_special(f1)
    f3=to_lower(f2)
    f4=rem_stopwords(f3)
    f5=stem_txt(f4)
    finaltxt.append(f5)   
```

```python
count=1
for rev in finaltxt :
    print(count,f".) ",rev,f"\n")
    count=count+1
```
